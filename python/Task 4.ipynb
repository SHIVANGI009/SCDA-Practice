```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Player': ['Anand', 'Anand', 'Vidit', 'Vidit', 'Anand'],
    'Opening': ['Ruy Lopez', 'Sicilian', 'Ruy Lopez', 'French', 'Sicilian'],
    'Accuracy': [92, 88, 85, 90, 95],
    'Moves': [45, 30, 50, 40, 60]
}

df= pd.DataFrame(data)
```

```python
df
```

```python
# Basic Grouping (Ek column par):
# Mujhe dekhna hai ki har player ki overall average (mean) accuracy kya hai:

avg_accuracy = df.groupby("Player")["Accuracy"].mean()
avg_accuracy
```

```python
# Advanced Grouping (Multiple calculations ek saath):
# Agar mujhe har player ka average accuracy BHI chahiye, aur total moves BHI chahiye, 
# toh hum .agg() (aggregation) function ka use karte hain.

adv_group = df.groupby("Player").agg({
    "Accuracy":"mean",
    "Moves":"sum"
})
print(adv_group)
```
