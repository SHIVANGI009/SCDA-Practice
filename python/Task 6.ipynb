```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [100, 150, 120, 200]
```

```python
# --- 1. LINE CHART ---
plt.plot(months,sales,color="Olive",marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

```python
# --- 2. BAR CHART ---
plt.bar(months,sales,color="Olive")
plt.title("Monthly Sales Comparison")
plt.show()
```

```python
# --- 3. PIE CHART ---
plt.pie(sales,labels=months,autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme(style="darkgrid")
df = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Grocery'],
    'Revenue': [5000, 3000, 4500]
})
# --- SEABORN BAR CHART ---
sns.barplot(x="Category", y="Revenue", data=df, hue="Category", palette="viridis", legend=False)
plt.title("Revenue by Category")
plt.show()
```
