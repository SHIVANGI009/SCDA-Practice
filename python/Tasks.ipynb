```python
# groupby ka use karke pata lagao ki har 'Genre' ki total kitni 'Copies_Sold' hui hain.

data = {
    'Genre': ['Fiction', 'Non-Fiction', 'Fiction', 'Sci-Fi', 'Non-Fiction'],
    'Book_Name': ['Book A', 'Book B', 'Book C', 'Book D', 'Book E'],
    'Copies_Sold': [50, 30, 45, 20, 60]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
copy_sold= df.groupby('Genre')['Copies_Sold'].count()
```

```python
copy_sold
```

```python
# Us total sales data ka ek Bar Chart plot karo (Graph ka title aur color tum apni marzi ka rakh sakti ho).

import matplotlib.pyplot as plt
```

```python
copy_sold.plot(kind='bar', color=['grey','black','blue'], title='Total Sales of Genre')

plt.ylabel('Total Books')
plt.show()
```
